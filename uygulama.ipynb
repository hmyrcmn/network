```python
import pandas as pd
import matplotlib.pyplot as plt

# Örnek veri
data = {'time': [0, 1, 2, 3, 4, 5], 'pdr': [100, 98, 97, 95, 93, 90], 'delay': [0, 1, 1.5, 2, 2.5, 3]}
df = pd.DataFrame(data)

# Packet Delivery Rate (PDR) grafiği
plt.figure(figsize=(10, 5))
plt.plot(df['time'], df['pdr'], label='PDR')
plt.xlabel('Time (s)')
plt.ylabel('PDR (%)')
plt.title('Packet Delivery Rate Over Time')
plt.legend()
plt.show()

# Gecikme grafiği
plt.figure(figsize=(10, 5))
plt.plot(df['time'], df['delay'], label='Delay', color='red')
plt.xlabel('Time (s)')
plt.ylabel('Delay (s)')
plt.title('Packet Delay Over Time')
plt.legend()
plt.show()
```
